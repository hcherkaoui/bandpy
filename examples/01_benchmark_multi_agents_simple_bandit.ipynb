Classical simple bandit agents benchmark
===================================

```python
import matplotlib.pyplot as plt
import numpy as np
from bandpy import launch_experiment
from bandpy.env import BernoulliKBandit, GaussianKBandit, CanonicalLinearBandit
from bandpy.controller import DecentralizedController
from bandpy.agents import FollowTheLeader, Uniform, EC, UCB
from bandpy.utils import check_random_state
from bandpy.plotting import plot_regrets
```

Global variables
-----------------------

```python
plt.style.use('tableau-colorblind10')
N_AGENTS = 10
MAX_RANDINT = 10000
n_trials = 20
T = 100
seed = None
n_jobs = 7
verbose = True
```

Envs/agents definition
--------------------------------

```python
env_names = ["Bernoulli-K-bandit",
             "Gaussian-K-Bandit",
             "Linear-Bandit-2D",
             ]
envs = [BernoulliKBandit(p=[0.1, 0.5, 0.6], T=T, seed=seed),
        GaussianKBandit(mu=[0.1, 0.5, 0.6], sigma=[1.0, 1.0, 1.0], T=T, seed=seed),
        CanonicalLinearBandit(d=2, delta=0.01, T=T, seed=seed),
        ]

controller_names = ["Follow-the-leader agent",
                    "Uniform agent",
                    "E-C agent",
                    "UCB agent",
                    ]
controllers = [DecentralizedController(N=N_AGENTS, agent_cls=FollowTheLeader, agent_kwargs={'K': 3, 'seed': seed}),
               DecentralizedController(N=N_AGENTS, agent_cls=Uniform, agent_kwargs={'K': 3, 'seed': seed}),
               DecentralizedController(N=N_AGENTS, agent_cls=EC, agent_kwargs={'K': 3, 'm': 0.2, 'T': T, 'seed': seed}),
               DecentralizedController(N=N_AGENTS, agent_cls=UCB, agent_kwargs={'K': 3, 'delta': 0.01, 'seed': seed}),
               ]
```

```python
print(f"[Main] Running controllers: {controller_names}.")
print(f"[Main] Environment: '{env_names}'.")
```

Running the experiment
----------------------------------

```python
rng = check_random_state(seed)
seeds = rng.randint(MAX_RANDINT, size=n_trials)
simulation_parameters = dict(env_names=env_names, envs=envs,
                             controller_names=controller_names,
                             controllers=controllers,
                             n_trials=n_trials, n_jobs=n_jobs,
                             seeds=seeds, enable_controller_early_stopping=False,
                             verbose=verbose)
all_results = launch_experiment(**simulation_parameters)
```

```python
all_regrets = dict()
for env_name, controllers in all_results.items():
    all_regrets[env_name] = dict()
    for controller_name, results in controllers.items():
        l_regrets = []
        for result in results:
            _, regrets, _ = result
            l_regrets.append(regrets)
        all_regrets[env_name][controller_name] = np.array(l_regrets)
```

```python
plot_regrets(all_regrets, fig_fname=None, verbose=verbose)
```
